```python
import numpy as np
import seaborn as sns
import pytreebank
from keras_preprocessing.text import text_to_word_sequence
from keras.preprocessing.sequence import pad_sequences
from bert_embedding import BertEmbedding
```

```python
sst_data = pytreebank.load_sst()
train = sst_data['train']
bert = BertEmbedding()
```

```python
x_list = []
y_list = []
for t in train:
    y, x = t.to_labeled_lines()[0]
    y_list.append(y)
    str_list, arr_list = zip(*bert([x]))
    #print(str_list)
    x_list.append(np.asarray(arr_list))
    print(x_list[-1].shape)
```

```python
x_list[0].dtype
```

```python
len_list = [i.shape[1] for i in x_list]
max_len = max(len_list)
min_len = min(len_list)
print(str(max_len), str(min_len))
```

```python
sns.set(style="whitegrid")
ax = sns.boxplot(data=len_list)
sns.plt.show()
```
